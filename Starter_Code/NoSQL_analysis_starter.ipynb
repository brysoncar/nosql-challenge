### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority, Using the LocalAuthorityCode
# 3. Sorts the matches from highest to lowest

pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityCode", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

pipeline_results = establishments_collection.aggregate(pipeline)

# I have to convert the pipeline_results into a list to count them and print them

pipeline_list = list(pipeline_results)
```

```python
# Print the number of documents in the result

result_count = len(pipeline_list)
print(result_count)
```

```python
# Print the first 10 results

for result in pipeline_list[:10]:
    print(result)
```

```python
# Convert the result to a Pandas DataFrame

pipeline_df = pd.DataFrame(pipeline_list)

# Display the number of rows in the DataFrame

pipeline_rows = pipeline_df.shape[0]
print(pipeline_rows)

# Display the first 10 rows of the DataFrame

pipeline_df.head()
```
